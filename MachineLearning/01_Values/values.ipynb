## Mean, Median, and Mode

In Machine Learning (and in mathematics) there are often three values that interests us:

### .Mean - The average value
### .Median - The mid point value
### .Mode - The most common value

```python
# Example: We have registered the speed of 13 cars:

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

mean_speed = sum(speed)/len(speed)
print(mean_speed)
```

## The NumPy module has a method for this.

```python
# Use the NumPy mean() method to find the average speed:
import numpy as np
mean_speed = np.mean(speed)
print(f"mean_speed: {mean_speed}")
```

```python
# Use the NumPy median() method to find the middle value:
import numpy as np
median_speed = np.median(speed)
print(f"median_speed: {median_speed}")
```

```python
# Use the SciPy mode() method to find the number that appears the most:
from scipy import stats
mode_speed = stats.mode(speed)
print(f"mode_speed: {mode_speed}")
```

# Standard Deviation?
Standard deviation is a number that describes how spread out the values are.

A low standard deviation means that most of the numbers are close to the mean (average) value.

A high standard deviation means that the values are spread out over a wider range.

```python
# Example: We have registered the speed of 13 cars:
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

mean_speed = np.mean(speed)
print(f"mean_speed: {mean_speed}")

# Calculate the standard deviation of the speed:
std_speed = np.std(speed)
print(f"std_speed: {std_speed}")
```

Meaning that most of the values are within the range of 9.26 from the mean value, which is 80.50.

As you can see, a higher standard deviation indicates that the values are spread out over a wider range.

##  Variance is another number that indicates how spread out the values are.

### In fact, if you take the square root of the variance, you get the standard deviation!

### the other way around, if you multiply the standard deviation by itself, you get the variance!

### To calculate the variance you have to do as follows:

```python
import numpy as np
from numpy import mean, median, std
speed = [32,111,138,28,59,77,97]
```

```python
# First Find mean value
mean_value = np.mean(speed)
print(f"mean_value: {mean_value}")
```

```python
# Calculate the variance of the speed:
variance = np.var(speed)
print(f"variance: {variance}")
```

## Steps to Calculate Variance

```python
# first find mean value
mean_value = np.sum(speed)/len(speed)
# For each value find difference from mean value, then square it
squared_diff = [(x - mean_value)**2 for x in speed]
for i in range(len(speed)):
    print(f"speed: {speed[i]}, mean_value: {mean_value}, squared_diff: {squared_diff[i]}")
    
# Then find the average of those squared differences    
squared_diff_sum = sum(squared_diff)
variance = squared_diff_sum/len(speed)
print(f"variance: {variance}")
```

##  Percentiles
Percentiles are used in statistics to give you a number that describes the value that a given percent of the values are lower than.

Example: Let's say we have an array that contains the ages of every person living on a street.

```python
# ages of 21 persons who attended a party:
# What is the 75. percentile ages of the persons?
ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]
percentile_75 = np.percentile(ages, 75)
print(f"75th percentile age: {percentile_75}, which means that 75% of the ages are below this value.")
```
