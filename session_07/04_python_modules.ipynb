##  What is a Module/Python Module/Program Module?
Consider a module to be the same as a code library.

A file containing a set of functions, variables and processes.
##  Create a Program Module
To create a module just save the code you want in a file with the file extension .py:

```python
def greeting(name):
  print("Hello, " + name)

# save it as mymodule.py
```

```python
import mymodule

mymodule.greeting("Jonathan")
```

##  Variables in Module
The module can contain functions, as already described, but also variables of all types (arrays, dictionaries, objects etc):

```python
person1 = {
  "name": "John",
  "age": 36,
  "country": "Norway"
}

# save it as mymodule.py
```

```python
import importlib
import mymodule
importlib.reload(mymodule) # reload the module to get the updated code
a = mymodule.person1["age"]
print(a)
```

##  Re-naming a Module
You can create an alias when you import a module, by using the as keyword:

```python
# processes 
import mymodule as mx # rename the module mymodule as mx

a = mx.person1["age"]
print(a)
```

##  Built-in Modules
There are several built-in modules in Python, which you can import whenever you like.

```python
import platform

x = dir(platform)
print(x)
```

##  Import From Module
You can choose to import only parts from a module, by using the from keyword.

```python
# mymodule.py
# functions and variables in mymodule.py
def greeting(name):
  print("Hello, " + name)

# dictionary in mymodule.py
person1 = {
  "name": "John",
  "age": 36,
  "country": "Norway"
}
```

```python
# import only the person1 dictionary from the module
from mymodule import person1
print (person1["age"])
```

## Examples of Embedded python modules


```python
# Datetime

# The datetime module has many methods to return information about the date object.

import datetime

x = datetime.datetime.now()
print(x)
print(x.year)
print(x.strftime("%A")) # Display the name of the weekday
```

##  Creating Date Objects
To create a date, we can use the datetime() class (constructor) of the datetime module.

The datetime() class requires three parameters to create a date: year, month, day.

```python
import datetime

x = datetime.datetime(2020, 5, 17) # Creating a date object

print(x)
```

##  The strftime() Method
The datetime object has a method for formatting date objects into readable strings.

The method is called strftime(), and takes one parameter, format, to specify the format of the returned string:

```python
import datetime

x = datetime.datetime(2018, 6, 1)

print(x.strftime("%B")) # Display the name of the month
```

## Built-in Math Functions


```python
# The min() and max() functions can be used to find the lowest or highest value in an iterable:

x = min(5, 10, 25)
y = max(5, 10, 25)

print(x)
print(y)
```

```python
# The abs() function returns the absolute (positive) value of the specified number:

x = abs(-7.25)

print(x)
```

```python
# The pow(x, y) function returns the value of x to the power of y (xy).

x = pow(4, 3)

print(x)
```

```python
import math

x = math.sqrt(64)

print(x)
```

```python
import math

x = math.ceil(1.4) # rounds a number upwards to its nearest integer
y = math.floor(1.4) # rounds a number downwards to its nearest integer

print(x) # returns 2
print(y) # returns 1
```

```python
# The math.pi constant, returns the value of PI (3.14...):

import math

x = math.pi

print(x)
```
