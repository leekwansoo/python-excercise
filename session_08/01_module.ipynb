## What is a Module

Consider a module to be the same as a code library.

A file containing a set of functions you want to include in your application.

## Create a Module

To create a module just save the code you want in a file with the file extension .py:

```python
# Create mymodule.py

def greeting(name):
    print("Hello, "+ name)
```

```python
# Test importing and using the module
import mymodule

# Use the function from the module
mymodule.greeting("Alice")
mymodule.greeting("World")
```

##  Variables in Module
The module can contain functions, as already described, but also variables of all types (arrays, dictionaries, objects etc):

```python
# Save this code in the file mymodule.py
person1 = {
  "name": "John",
  "age": 36,
  "country": "Norway"
}

print(person1)
print(person1["name"])
```

```python
# Simple version - no importlib needed in normal cases!
import mymodule

# Access the person1 dictionary
a = mymodule.person1["age"]
print("Age:", a)
print("Name:", mymodule.person1["name"])
print("Country:", mymodule.person1["country"])
```

```python
# Simple version - the code you originally wanted to run
import mymodule

a = mymodule.person1["age"] 
print(a)
```

```python
# Fresh start - no importlib needed!
# (This works because mymodule.py now contains person1)

import mymodule

# Access everything directly
print("From mymodule:")
print(f"Age: {mymodule.person1['age']}")
print(f"Name: {mymodule.person1['name']}")

# Use the greeting function too
mymodule.greeting("Module User")
```

```python
# WHEN YOU DO NEED importlib.reload():
# Only when you modify a module AFTER importing it

# Example scenario:
# 1. import mymodule        # Works fine
# 2. Edit mymodule.py       # Add new content to the file
# 3. import mymodule        # Still uses OLD cached version! 
# 4. Need: importlib.reload(mymodule)  # Forces re-read

print("Summary:")
print("✅ Normal use: Just 'import mymodule'")
print("⚡ After editing: Need 'importlib.reload(mymodule)'") 
print("🔄 Fresh session: Just 'import mymodule' again")
```

```python
# Create yourmodule.py using Python file operations
content = '''def greeting(name):
    print("Hello, " + name)'''

# Write the content to a new file
with open('yourmodule.py', 'w') as file:
    file.write(content)

print("File 'yourmodule.py' has been created successfully!")
```

```python
# Test the newly created yourmodule.py
import yourmodule

yourmodule.greeting("Python Learner")
yourmodule.greeting("VS Code User")
```
